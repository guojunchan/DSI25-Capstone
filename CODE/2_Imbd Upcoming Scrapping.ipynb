```python
#importing the libraries needed 
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
from time import sleep
from random import randint

headers = {"Accept-Language": "en-US,en;q=0.5"}

#declaring the list of empty variables, So that we can append the data overall
movie_name = []
year = []
time=[]
rating=[]
metascore =[]
votes = []
gross = []
description = []
Director = []
Stars = []
mgenre = []

#creating an array of values and passing it in the url for dynamic webpages
pages = np.arange(np.datetime64('2021-12'), 13)

#the whole core of the script
for page in pages:
    page = requests.get("https://www.imdb.com/movies-coming-soon/"+str(page)+"/?ref_=cs_dt_nx")
    soup = BeautifulSoup(page.text, 'html.parser')
    movie_data = soup.findAll('div', attrs = {'class': 'list_item odd' })
    sleep(randint(2,8))
    for store in movie_data:
        name = store.h4.a.text
        movie_name.append(name)

    
        genre = store.find_all('span')
        mgenre.append(genre)

     
#creating a dataframe 
movie_list = pd.DataFrame({ "Movie Name": movie_name,"Genre": mgenre})
```

```python
movie_list.to_csv("../data/Coming soon.csv")
```
